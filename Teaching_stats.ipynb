```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as graph
import seaborn as sns
import scipy.stats as stats
```

```python
a = np.random.normal(100, 15, 100)
b = np.random.normal(130, 5, 100)
```

```python
graph.figure(figsize=(12, 5))
sns.distplot(a, label='a')
sns.distplot(b, label='b')
# graph.axvline(100)
graph.legend()
graph.show()
```

```python
stats.ttest_ind(a, b)
```

## Correlations

```python
from rpy2.robjects import r, pandas2ri
from pydataset import data
```

```python
cars_data = data('mtcars')
cars_data['car'] = cars_data.index
display(cars_data.head())
```

```python
sns.jointplot(cars_data['mpg'], cars_data['wt'], kind='reg')
graph.xlabel('Miles/(US) gallon', fontsize=15)
graph.ylabel('Weight (1000 lbs)', fontsize=15)
graph.show()
```

```python
sns.jointplot(cars_data['mpg'], cars_data['hp'], kind='reg')
graph.xlabel('Miles/(US) gallon', fontsize=15)
graph.ylabel('Horsepower', fontsize=15)
graph.show()
```
